### One to many relationship

```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

1.	Report the account representative for each customer.

```python
run("""
SELECT 
    customerName, CONCAT(firstname," ",lastname) AS employeeName
FROM
    customers 
JOIN 
    employees ON salesRepEmployeeNumber = employeeNumber
""")
```

2.	Report total payments for Atelier graphique.

```python
run("""
SELECT 
    SUM(amount) AS TotalPay 
FROM 
    customers 
JOIN 
    payments USING(customerNumber) 
WHERE 
    customerName = 'Atelier Graphique'
""")
```

3.	Report the total payments by date

```python
run("""
SELECT
    paymentDate, SUM(amount) AS Total
FROM 
    payments 
GROUP BY
    paymentDate
""")
```

4.	Report the products that have not been sold.

```python
run("""
SELECT
    productCode, productName, orderNumber
FROM 
    products
LEFT JOIN
    orderDetails USING (productCode)
WHERE
    orderNumber IS NULL
""")
```

5.	List the amount paid by each customer.

```python
run("""
SELECT 
    customerNumber, customerName, SUM(amount) AS TotalPay 
FROM 
    customers 
JOIN 
    payments USING(customerNumber) 
GROUP BY 
    customerNumber
""")
```

6.	How many orders have been placed by Herkku Gifts?

```python
run("""
SELECT 
    COUNT(*) AS totalOrder 
FROM 
    customers 
JOIN 
    orders USING (customerNumber) 
WHERE 
    customerName = 'Herkku Gifts'
""")
```

7.	Who are the employees in Boston?

```python
run("""
SELECT
    CONCAT(firstName," ",lastName) AS employeeName
FROM 
    employees
JOIN
    offices USING(officeCode)
WHERE
    city = 'Boston'
""")
```

8.	Report those payments greater than \$100,000. Sort the report so the customer who made the highest payment appears first.

```python
run("""
SELECT 
    customerName, amount
FROM
    customers
JOIN
    payments USING (customerNumber) 
WHERE 
    amount>100000
ORDER BY 
    amount DESC
""")
```

9.	List the value of 'On Hold' orders.

```python
run("""
SELECT orderNumber, SUM(quantityOrdered * priceEach) AS value
FROM
    orders
JOIN
    orderdetails USING (orderNumber)
WHERE
    status = 'On Hold'
GROUP BY 
    orderNumber
""")
```

10.	Report the number of orders 'On Hold' for each customer.

```python
run("""
SELECT 
    customerName, COUNT(*) AS statOnHold
FROM
    customers
JOIN
    orders USING (customerNumber)
WHERE
    status = 'On Hold'
GROUP BY
    customerNumber
""")
```

### Many to many relationship

1.	List products sold by order date.

```python
run("""
SELECT
    orderDate, productName 
FROM
    products
JOIN
    orderdetails USING (productCode)
JOIN
    orders USING (orderNumber)
ORDER BY 
    orderDate
""")
```

2.	List the order dates in descending order for orders for the 1940 Ford Pickup Truck.

```python
run("""
SELECT
    orderNumber, orderDate 
FROM
    products
JOIN
    orderdetails USING (productCode)
JOIN
    orders USING (orderNumber)
WHERE 
    productName = '1940 Ford Pickup Truck'
ORDER BY 
    orderDate DESC 
""")
```

3.	List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?

```python
run("""
SELECT
    customerName, orderNumber, SUM(quantityordered*priceEach) AS orderValue
FROM
    customers
JOIN
    orders USING (customerNumber)
JOIN
    orderdetails USING (orderNumber)
GROUP BY
    orderNumber
HAVING 
    orderValue>25000
""")
```

4.	Are there any products that appear on all orders?

```python
run("""
SELECT
    productCode,COUNT(*) AS number
FROM
    orderdetails
GROUP BY
    productCode HAVING number = (SELECT COUNT(*) FROM orders)
""")
```

5.	List the names of products sold at less than 80% of the MSRP.

```python
run("""
SELECT
    orderNumber, productName, priceEach, MSRP
FROM
    products
JOIN
    orderdetails USING (productCode)
JOIN
    orders USING (orderNumber)
WHERE 
    priceEach < (80/100)*MSRP
""")
```

6.	Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)

```python
run("""
SELECT
    productName, priceEach-buyPrice AS profit, buyPrice
FROM
    products
JOIN
    orderdetails USING (productCode)
WHERE
    priceEach-buyPrice > buyPrice    
""")
```

7.	List the products ordered on a Monday.

```python
run("""
SELECT
    DAYNAME(orderDate), productName
FROM
    products
JOIN
    orderdetails USING (productCode)
JOIN
    orders USING (orderNumber)
WHERE 
    DAYOFWEEK(orderDate) = 2
""")
```

8.	What is the quantity on hand for products listed on 'On Hold' orders?

```python
run("""
SELECT 
    SUM(quantityOrdered) AS total
FROM
    orders 
JOIN 
    orderdetails USING (orderNumber)
Where
    status='On Hold'
""")
```
